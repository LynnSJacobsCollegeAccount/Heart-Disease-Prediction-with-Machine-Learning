```python
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import pandas as pd
```

```python
heartData = pd.read_csv("df_clean.csv")
X = heartData.drop(columns=["TenYearCHD"])
y = heartData["TenYearCHD"]
```

### Splitting and scaling:
Outputs X_train_scaled, X_test_scaled, y_train, y_test

Standard scaling the continuous features

```python
# STEP 1: Initial Split with stratification:
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3932, test_size=0.2, stratify=y)

# STEP 2: Standard scale the continuous columns
cont_cols = ["age", "cigsPerDay", "totChol", "sysBP", "diaBP", "BMI", "heartRate", "glucose"]
scaler = StandardScaler()

# Copy to avoid modifying original data
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Fit on train
scaler.fit(X_train[cont_cols])

# Transform
X_train_scaled.loc[:, cont_cols] = scaler.transform(X_train[cont_cols])
X_test_scaled.loc[:, cont_cols] = scaler.transform(X_test[cont_cols])
```

### Rest of the Logistic Regression Code: 
Using the procedure from the sling academy article saved in my computer browser favorites. (Article name: Logistic Regression with Cross-Validation in Scikit-Learn)

```python
#Initialize the model
model = LogisticRegression(max_iter=200)
# Evaluate using 10-fold cross-validation
evaluations = cross_val_score(model, X_train_scaled, y_train, cv=10, scoring='f1')
print('Cross-validation f1 scores:', evaluations)
print('Mean cross-validation f1 score:', np.mean(evaluations))

evaluations = cross_val_score(model, X_train_scaled, y_train, cv=10, scoring='average_precision')
#print('Cross-validation scores:', evaluations)
print('Mean cross-validation PR_AUC score:', np.mean(evaluations))


# Fit the model
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)
#print('Predicted values:', y_pred)
```
