```python
df_clean = df.copy()
```

```python
#Imputing missing values with median and mode.

# fill education with mode
df_clean['education'] = df_clean['education'].fillna(df_clean['education'].mode()[0])

# fill BPMeds with mode
df_clean['BPMeds'] = df_clean['BPMeds'].fillna(df_clean['BPMeds'].mode()[0])

# fill continuous columns with median
for col in ['cigsPerDay', 'totChol', 'BMI', 'heartRate', 'glucose']:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
#Changing discrete variables to type int.

for col in ['male', 'education', 'currentSmoker', 'BPMeds',
            'prevalentStroke', 'prevalentHyp', 'diabetes', 'TenYearCHD']:
    df_clean[col] = df_clean[col].astype(int)
```

```python
df_clean.head()
```

```python
#Checking to see if binary variables have any unintended values.

for col in ['male', 'currentSmoker', 'BPMeds', 'prevalentStroke',
            'prevalentHyp', 'diabetes', 'TenYearCHD']:
    print(col, df_clean[col].unique())
```

```python
df_clean['education'].unique()
```

```python
#Checking for unusual outliers

for col in ['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']:
    print(col, df_clean[col].min(), df_clean[col].max())
```

```python
#Checking for outliers by visualizing a box plot.

for col in ['cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']:
    plt.figure()
    df_clean.boxplot(column=col)
    plt.title(col)
    plt.show()
```
